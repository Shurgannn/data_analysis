```python
#увеличим дефолтный размер графиков
from pylab import rcParams
rcParams['figure.figsize'] = 12,9

# импортируем нужные библиотеки
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = 'C:/My documents/names'
```

```python
# Построить график изменения количества имен Ruth и Robert с 1900 по 2000.
names_list = ['Ruth', 'Robert']
Ruth_list = []
Robert_list = []
for name in names_list:
    y = 1900
    while y < 2001:
        names = pd.read_csv(
            f'{file_path}/yob{y}.txt',
            names=['Name', 'Gender', 'Count'])
        names_barh = names[names['Name']==name]
        for i in names_barh['Count']:
            if name == 'Ruth':
                Ruth_list.append(i)
            if name == 'Robert':
                Robert_list.append(i)                
        y += 1
plt.plot(Ruth_list)
plt.plot(Robert_list)
```
